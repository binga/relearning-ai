```python
model = GLiNER.from_pretrained("urchade/gliner_small-v2.1")
```

```python
n_runs = 30
```

```python
# cold start run

entities = model.predict_entities(text, labels)

times = []
# warm start run
for _ in range(n_runs):
    start_time = time.time()
    entities = model.predict_entities(text, labels)
    end_time = time.time()
    times.append(end_time - start_time)

latency1 = sum(times) / n_runs
print(f"Average run duration: {latency1}")
```

```python
model = GLiNER.from_pretrained("urchade/gliner_small-v2.1", _attn_implementation="eager")
```

```python
# cold start run

entities = model.predict_entities(text, labels)

times = []
# warm start run
for _ in range(n_runs):
    start_time = time.time()
    entities = model.predict_entities(text, labels)
    end_time = time.time()
    times.append(end_time - start_time)

latency2 = sum(times) / n_runs
print(f"Average run duration: {latency2}")
```

```python
entities = model.predict_entities(text, labels)
for entity in entities:
    print(entity["text"], "=>", entity["label"])
```

```python
speedup = (latency1 - latency2) * 100/latency1
print(f"Achieved speedup with flashdeberta: {round(speedup, 2): .2f}%")
```
